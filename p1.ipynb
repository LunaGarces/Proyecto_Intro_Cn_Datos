```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
encuesta = pd.read_spss('Datasets/principal-encuestas/VIF_Final.sav')
encuesta.head()
```

```python
ingresos = pd.read_spss('Datasets/ingresos/Ingresos_Final.sav')
ingresos.head()
```

```python
encuesta.columns
cols_e = ['VIOLENCIA_PUBLICA',
       'VIOLENCIA_PUBLICA_NINEZ', 'VILENCIA_PUBLICA_VIDA',
       'VIOLENCIA_LABORAL_ANIO', 'VIOLENCIA_LABORAL_NINEZ',
       'VIOLENCIA_LABORAL_VIDA','VIOLENCIA_ECONOMICA', 'DEPENDENCIA_ECONOMICA']
encuesta[cols_e]
```

Las columnas elegidas están compuestas de datos 
#### 1 : SI 
#### 2 : NO

#### encuesta['DEPENDENCIA_ECONOMICA'] 
- 0 Baja
- 1 Mediana 
- 2 Alta 
#### GRUPO_SOCIOECONOMICO: 
- 1 : Alto 
- 2 : Medio 
- 3 : Bajo 


```python
for i in cols_e:
    sns.set_style('whitegrid')
    g = sns.catplot(x= i, hue="REG_REC", col="GRUPO_SOCIOECONOMICO",
                    data=encuesta,
                    kind="count", height=30, aspect=.2, palette='pastel')
                    
    plt.legend(fontsize = 20, loc = 'upper left')
    for ax in g.axes.flat[1:]:
        sns.despine(ax=ax, left=True)

    for ax in g.axes.flat:
        ax.set_xlabel(ax.get_title(), fontsize = 15)
        ax.set_title('')
        ax.margins(x=0.1) # slightly more margin as a separation
    plt.subplots_adjust(wspace=0, bottom=0.18, left=0.06)
    plt.show()
```

```python
enc_reg = encuesta.groupby(by= ['REG_REC', 'GRUPO_SOCIOECONOMICO'])[cols_e].agg(pd.Series.median).reset_index().loc[:, ['REG_REC', 'GRUPO_SOCIOECONOMICO', 
                            'VIOLENCIA_PUBLICA','VIOLENCIA_PUBLICA_NINEZ', 
                            'VILENCIA_PUBLICA_VIDA','VIOLENCIA_LABORAL_ANIO', 
                            'VIOLENCIA_LABORAL_NINEZ','VIOLENCIA_LABORAL_VIDA',
                            'VIOLENCIA_ECONOMICA', 'DEPENDENCIA_ECONOMICA']]
enc_reg
```

```python
ing_reg = ingresos.groupby(by = ['CODIGO_REGION', 'TIPO_ESTRATO_HOGAR']).mean().reset_index().loc[:, ['CODIGO_REGION', 'TIPO_ESTRATO_HOGAR','MONTO_HOGAR']]
ing_reg
```

#### 'TIPO_ESTRATO_HOGAR': 
- 1 : Ciudad
- 2 : Resto de Área Urbana 
- 3 : Rural 

```python
sns.barplot(y = 'MONTO_HOGAR', x ='CODIGO_REGION', hue='TIPO_ESTRATO_HOGAR', data= ing_reg, palette='pastel')
```

```python
data = pd.concat([enc_reg, ing_reg], axis=1, join='inner')
data
```
